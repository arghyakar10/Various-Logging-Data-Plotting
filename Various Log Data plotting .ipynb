```python
import lasio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('15_9-19A-CORE.csv')
```

```python
df.head()
```

```python
df.plot(subplots=True,figsize=(9,12))
```

```python
plt.figure(figsize=(4,16))
plt.style.use('default')

fig,ax=plt.subplots(nrows=1,ncols=2)

fig.set_figheight(15)
fig.set_figwidth(15)

ax[0].scatter(df['CPOR'],df['DEPTH'])
ax[1].scatter(np.log10(df['CKHG']),df['DEPTH'],color='red')


ax[0].set_title('Core Porosity')
ax[1].set_title('Core Permeability')


ax[0].set_ylim(4000,3840)
ax[1].set_ylim(4010,3825)

ax[0].grid()
ax[1].grid()
plt.show()
```

```python
plt.style.use('default')
plt.scatter(df['CPOR'],df['CKHG'],color='purple')
plt.semilogy()
plt.title('Poro-Perm')
plt.ylim(0.01,10000)
plt.xlim(0,50)
plt.grid()
plt.show()
```

```python
plt.style.use('default')
plt.hist(df['CPOR'],bins=30,color='green',edgecolor='black')

plt.title('Core Porosity')
plt.show()
```

```python
plt.style.use('default')
plt.hist(df['CGD'],bins=30,color='pink',edgecolor='black')
plt.title('Core grain Density')
plt.show()
```
